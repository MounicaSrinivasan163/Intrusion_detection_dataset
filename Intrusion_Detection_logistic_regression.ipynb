## 📘 LogisticRegression Parameters Explained (scikit-learn)



Below are the most commonly used parameters for `LogisticRegression` from `sklearn.linear_model`. Each point includes the parameter name, type, default value, and its usage.

---

 1. `penalty`
- **Type**: `str`, default = `'l2'`
- **Description**: Type of regularization to apply.
- **Options**: `'l1'`, `'l2'`, `'elasticnet'`, `'none'`
- `'l1'`: Lasso (sparse model)  
- `'l2'`: Ridge (shrinks coefficients)  
- `'elasticnet'`: Mix of L1 and L2  
- `'none'`: No regularization

---

### 2. `dual`
- **Type**: `bool`, default = `False`
- **Used only when** `solver='liblinear'` and `penalty='l2'`
- Solves the dual form of optimization problem (useful when n_samples > n_features)

---

### 3. `tol`
- **Type**: `float`, default = `1e-4`
- **Description**: Tolerance for stopping criteria
- Lower `tol` = more precise but longer training time

---

### 4. `C`
- **Type**: `float`, default = `1.0`
- **Description**: Inverse of regularization strength
- Smaller `C` → Stronger regularization  
- Larger `C` → Weaker regularization (model fits training data more)

---

### 5. `fit_intercept`
- **Type**: `bool`, default = `True`
- **Description**: Whether to add an intercept (bias) to the model
- Set to `False` if data is already centered

---

### 6. `intercept_scaling`
- **Type**: `float`, default = `1`
- **Used only when** `solver='liblinear'` and `fit_intercept=True`
- **Purpose**: Scales the synthetic feature added for intercept optimization

---

### 7. `class_weight`
- **Type**: `dict`, `'balanced'`, or `None`; default = `None`
- **Purpose**: Handles class imbalance
- `'balanced'`: Auto-adjusts weights inversely proportional to class frequencies
- Custom: `{0: 1, 1: 3}` gives class 1 triple the importance

---

### 8. `random_state`
- **Type**: `int`, `RandomState`, or `None`; default = `None`
- **Purpose**: Controls randomness (e.g. for data shuffling in solvers)
- Set for reproducibility (e.g. `random_state=42`)

---

### 9. `solver`
- **Type**: `str`, default = `'lbfgs'`
- **Description**: Optimization algorithm to use
- **Options**:
  - `'liblinear'`: Good for small datasets, supports L1
  - `'lbfgs'`: Fast, supports multinomial loss
  - `'sag'`: Good for large datasets (requires standardized features)
  - `'saga'`: Supports all penalties + large data
  - `'newton-cg'`: Similar to lbfgs for L2

---

### 10. `max_iter`
- **Type**: `int`, default = `100`
- **Description**: Maximum number of iterations to converge
- Increase if convergence warning occurs

---

### 11. `multi_class`
- **Type**: `str`, default = `'auto'`
- **Options**:
  - `'auto'`: Chooses `'ovr'` or `'multinomial'` based on solver
  - `'ovr'`: One-vs-Rest strategy (default for `liblinear`)
  - `'multinomial'`: True softmax regression

---

### 12. `verbose`
- **Type**: `int`, default = `0`
- Controls the verbosity of the solver
- Set to `1` or higher to print convergence messages

---

### 13. `warm_start`
- **Type**: `bool`, default = `False`
- If `True`, reuse solution from previous `.fit()` calls
- Useful for incremental or online learning

---

### 14. `n_jobs`
- **Type**: `int`, default = `None`
- **Purpose**: Number of CPU cores to use
- Only effective for solvers: `'liblinear'`, `'sag'`, `'saga'`
- Use `n_jobs=-1` to use all available cores

---

### 15. `l1_ratio`
- **Type**: `float`, default = `None`
- Used only when `penalty='elasticnet'`
- Value in `[0, 1]`:
  - `0` = L2 penalty only
  - `1` = L1 penalty only
  - In between = mixture of L1 and L2


## Hyperparameter Tuning

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.linear_model import LogisticRegression
from scipy.stats import uniform

# Define model
model = LogisticRegression()

# Define hyperparameter distributions
param_distributions = {
    'C': uniform(0.01, 10),  # Uniform distribution between 0.01 and 10
    'penalty': ['l2'],
    'solver': ['lbfgs', 'liblinear']
}

# Create RandomizedSearchCV
random_search = RandomizedSearchCV(
    estimator=model,
    param_distributions=param_distributions,
    n_iter=10,                 # Number of different parameter settings to try
    scoring='accuracy',        # Metric to optimize
    cv=5,                      # 5-fold cross-validation
    verbose=1,
    random_state=42,
    n_jobs=-1
)

# Fit to training data
random_search.fit(x_train, y_train)

# Best model and params
print(random_search.best_estimator_)
print(random_search.best_params_)
print(random_search.best_score_)
```

| Parameter             | Description                                                                                     |
| --------------------- | ----------------------------------------------------------------------------------------------- |
| `estimator`           | The ML model to tune (e.g., `LogisticRegression()`)                                             |
| `param_distributions` | Dictionary with parameter names as keys and distributions or lists of parameters to sample from |
| `n_iter`              | Number of parameter settings sampled (controls speed vs coverage)                               |
| `scoring`             | Metric to evaluate models (e.g. `'accuracy'`, `'roc_auc'`)                                      |
| `cv`                  | Cross-validation strategy (e.g. `5` = 5-fold CV)                                                |
| `verbose`             | Controls logging: `0` (silent), `1` (minimal), `2` (detailed)                                   |
| `random_state`        | Ensures reproducibility                                                                         |
| `n_jobs`              | Number of parallel jobs (e.g. `-1` for all CPUs)                                                |


```python
final_model = LogisticRegression(**random_search.best_params_)
final_model.fit(x_train, y_train)
y_pred = final_model.predict(x_test)

accuracy = accuracy_score(y_test, y_pred)
print("Test Accuracy:", accuracy)
```
